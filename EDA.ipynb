DATA **CLEANING**

```python
#Data Cleaning
df["Condition"].replace(["", "None", "NaN","Nan", "NULL"], np.nan, inplace=True)
df["Condition"] = df["Condition"].fillna("Not Bad")

df
```

```python
#df.drop("Condition",axis=1,inplace=True)
#df
```

```python
df["Area"].mean()
```

```python
df['Area'].fillna(df['Area'].mean(),inplace=True)
df
```

```python
#Fix eror
df["YearBuilt"] = df["YearBuilt"].replace(2050,2020)
df
```

Data transformation

```python
#Data Transformation
df["Area"] = df["Area"].astype(int)
df
```

```python
#Add column
df["PricePerArea"] = (df["Price"] / df["Area"]).map(lambda x: f"{x:.2f}")
# or this code =>  df["PricePerArea"] = (df["Price"] / df["Area"]).round(2) for round the float value
df
```

Group by:It is a powerful method used to group data based on one or more columns and perform aggregate on each group.

```python
#Data aggregation :Mean,Median,Max,Min,Count,sumui
grouped_data=df.groupby("Condition")["Price"].mean()
grouped_data
```

```python
df.groupby("Condition")["Price"].agg(["mean","median","max","min","count","sum"])
```

```python
#need to vizualize price less then 400000 lakhs and the condition is Good
df_filtered = df[(df["Price"]<400000) & (df["Condition"]=="Good")]
display(df_filtered['Id'])
```

Data Vizualization

```python
#Bar chart => used for Catogorical valu + Numerical Colun
plt.figure(figsize=(4, 3))
sns.countplot(x="Condition",data=df,color='y',edgecolor='black')
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.xlabel("Condition")
plt.title("Condition Count")
plt.show()
```

```python
#Line plot = used for continuous values
plt.figure(figsize=(3, 2))
sns.lineplot(x="Area", y="Price", data=df)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
#Scatter plot = see the relationship between continous values
plt.figure(figsize=(3, 2))
sns.scatterplot(x="Area", y="Price", data=df)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
#Histogram = to analyze the single column like Numeric
plt.figure(figsize=(3, 2))
sns.histplot(df["Price"], bins=10, kde=True)
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```
